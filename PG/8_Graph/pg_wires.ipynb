```python
def solution(n, wires):
    answer = n
    for sub in (wires[i+1:] + wires[:i] for i in range(len(wires))):
        s = set(sub[0])
        [s.update(v) for _ in sub for v in sub if set(v) & s] #  & :  ,    update :   
        answer = min(answer, abs(2 * len(s) - n))
    return answer
```
