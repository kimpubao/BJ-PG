N×N 크기의 공간에 물고기 M마리와 아기 상어 1마리가 있다. 공간은 1×1 크기의 정사각형 칸으로 나누어져 있다. 한 칸에는 물고기가 최대 1마리 존재한다.


아기 상어와 물고기는 모두 크기를 가지고 있고, 이 크기는 자연수이다. 가장 처음에 아기 상어의 크기는 2이고, 아기 상어는 1초에 상하좌우로 인접한 한 칸씩 이동한다.


아기 상어는 자신의 크기보다 큰 물고기가 있는 칸은 지나갈 수 없고, 나머지 칸은 모두 지나갈 수 있다. 아기 상어는 자신의 크기보다 작은 물고기만 먹을 수 있다. 따라서, 크기가 같은 물고기는 먹을 수 없지만, 그 물고기가 있는 칸은 지나갈 수 있다.


아기 상어가 어디로 이동할지 결정하는 방법은 아래와 같다.


- 더 이상 먹을 수 있는 물고기가 공간에 없다면 아기 상어는 엄마 상어에게 도움을 요청한다.
- 먹을 수 있는 물고기가 1마리라면, 그 물고기를 먹으러 간다.

# 구현 시 주의할 점
- BFS 탐색에서는 이동할 수 있는 칸과 먹을 수 있는 칸을 구분해서 처리해야 한다.
  - 이동 조건: 현재 칸의 물고기 크기 <= 상어 크기
  - 먹이 조건: 현재 칸의 물고기 크기 < 상어 크기
- 먹을 수 있는 물고기는 리스트에 (거리, x, y) 형식으로 저장한 뒤 정렬해 우선순위를 결정한다.
- 먹이를 먹을 때마다 상어 위치, 크기, 먹은 수, 시간 등을 반드시 갱신해야 한다.
- BFS마다 visited 배열을 새로 만들어줘야 한다 (매번 새로운 탐색이므로).


### BFS
- **Breadth-First Search(너비 우선 탐색)**의 줄임말
- BFS는 말 그대로 “가까운 곳부터 차근차근 넓게 퍼지듯이 탐색하는 방식
  - 지금 있는 위치에서 한 칸씩 갈 수 있는 모든 방향을 먼저 다 살펴보고,
  - 그 다음으로 두 칸 떨어진 곳,
  - 그리고 세 칸... 이런 식으로 거리순으로 탐색

### BFS의 작동 방식 (2차원 맵 기준)
1. 시작 위치를 큐(queue)에 넣는다.
2. 큐에서 하나를 꺼낸다.
3. 그 위치에서 갈 수 있는 칸(상하좌우)을 모두 큐에 넣는다.
4. 방문 표시를 해서 같은 곳을 중복해서 탐색하지 않도록 한다.
5. 큐가 빌 때까지 2~4번을 반복한다.

### BFS를 왜 쓰나요?
- 최단 거리를 찾고 싶을 때 좋아요!
  → DFS(깊이 우선 탐색)는 먼저 깊이 파고들기 때문에 최단 거리와는 거리가 있음

- 특히 아기 상어 문제처럼:
  - 가장 가까운 먹잇감 탐색
  - 갈 수 있는 모든 칸에서 조건 확인
  → 이런 상황에 BFS가 딱 적합

# 예시 코드(기초적인 BFS)

````python
from collections import deque

# 예: 5x5 맵에서 (0,0)에서 (4,4)까지 최단 거리 탐색
N = 5
grid = [[0]*N for _ in range(N)]
visited = [[False]*N for _ in range(N)]

dx = [-1, 1, 0, 0]  # 상하좌우
dy = [0, 0, -1, 1]

def bfs(x, y):
    q = deque()
    q.append((x, y, 0))  # 위치 + 거리
    visited[x][y] = True

    while q:
        x, y, dist = q.popleft()

        if x == 4 and y == 4:
            return dist  # 목표에 도달하면 거리 반환

        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]

            if 0 <= nx < N and 0 <= ny < N:
                if not visited[nx][ny] and grid[nx][ny] == 0:
                    visited[nx][ny] = True
                    q.append((nx, ny, dist + 1))
````

# 정답 코드

````python
from collections import deque

# 입력 처리
N = int(input())
board = [list(map(int, input().split())) for _ in range (N)]

# 상어의 시작 위치 찾기
for i in range(N):
    for j in range(N):
        if board[i][j] == 9:
            shark_x, shark_y = i,j
            board[i][j] = 0 # 상어가 있던 자리는 0으로 바꿈

# 전역 변수 초기화
shark_size = 2
eat_count = 0
total_time = 0

# 상하좌우 이동 방향
dx = [-1,0,0,1]
dy = [0,-1,1,0]

def bfs(x,y, size):
    visited = [[False]*N for _ in range(N)]
    q = deque()
    q.append((x,y,0))
    visited[x][y] = True
    fishes = []

    while q:
        cx, cy, dist = q.popleft()
        for i in range(4):
            nx, ny = cx + dx[i], dy[i]

            if 0 <= nx < N and 0 <= ny < N and not visited[nx][ny]:
                # 크기보다 큰 물고기가 있으면 못 감
                if board[nx][ny] > size:
                    continue
                visited[nx][ny] = True
                # 먹을 수 있는 물고기라면
                if 0 < board[nx][ny] < size:
                    fishes.append((dist+1, nx, ny))
                q.append((nx,ny, dist+1))
    
    # 거리, 위쪽, 왼쪽 우선순위로 정렬
    fishes.sort()
    return fishes[0] if fishes else None

# 시뮬레이션
while True:
    target = bfs(shark_x, shark_y, shark_size)

    if target is None:
        break # 더 이상 먹을 수 있는 물고기가 없음

    dist, nx, ny = target
    total_time += dist
    eat_count += 1
    board[nx][ny] = 0
    shark_x, shark_y = nx, ny

    if eat_count == shark_size:
        shark_size += 1
        eat_count = 0

# 출력
print(total_time)
````

## 📚 문제 요약

## 🔍 BFS 개념 (내 기준에서 정리)

- 완벽히 알진 못하지만, 대충 이런 느낌:
  - **가까운 곳부터 하나씩 탐색해 나가는 방식**
  - **Queue(큐)**를 사용해서 순서대로 꺼내가며 퍼져 나간다.
- BFS는 주로 **최단 거리** 문제나, **퍼지는 방식**이 필요할 때 쓰인다고 한다.
- 이번 문제에서는 가장 가까운 물고기를 찾기 위해 BFS를 사용함.

2. 거리 + 위치 우선순위로 정렬해서 가장 먼저 먹어야 할 물고기를 결정한다.
3. 상어가 이동하고, 시간도 더해주고, 물고기 수를 누적한다.
4. 누적된 수가 상어 크기만큼 쌓이면 크기를 키운다.
5. 더 이상 먹을 게 없으면 종료.
